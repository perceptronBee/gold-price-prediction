```python
columns_to_drop=["us_rates_%","CPI","GDP"]
df = df.drop(columns=columns_to_drop)
```

```python
dfo = df.filter(regex="open")
dfo
```

```python
df.columns
```

```python
dfo = pd.concat([dfo,df[['usd_chf', 'eur_usd','date']]],axis=1)
```

```python
dfo.tail()
```

```python
for i,nan in enumerate(dfo.isna().sum(axis=1)):
    if nan>6:
        dfo.drop(i,axis=0,inplace=True)
```

```python
dfo.reset_index()
```

```python
dfo.isna().sum()
```

```python
dfo["usd_chf"] = dfo["usd_chf"].bfill().ffill()
dfo["eur_usd"] = dfo["eur_usd"].bfill().ffill()
```

```python
dfo.isna().sum()
```

```python
dfo['date'] = pd.to_datetime(dfo['date'])
dfo['date'] = dfo['date'].astype('int64') // 10**9
```

```python
dfo.head()
```

```python
X=dfo.drop("gold open",axis=1)
y=dfo["gold open"]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from lazypredict.Supervised import LazyRegressor
lazy = LazyRegressor()
lazy.fit(X_train, X_test, y_train, y_test)
```
